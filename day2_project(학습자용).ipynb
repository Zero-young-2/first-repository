### step 1

```python
car_df[car_df[['Engine', 'Emission Class', 'Service history']].isna().all(axis=1)]
```

```python
car_df.info()
```

```python
car_df = car_df[~car_df[['Engine', 'Emission Class', 'Service history']].isna().all(axis=1)]
```

```python
car_df
```

```python
car_df.loc[:, 'title_F'] = car_df['title'].str.split().str[0]
```

```python
car_df.loc[:, 'title_E'] = car_df['title'].str.split().str[1]
```

```python
car_df.head()
```

```python
brand_df['title'] = brand_df['title'].str.upper()
brand_df.head()
```

```python
brand_df.rename(columns={'title':'title_F'}, inplace=True)
brand_df.head()
```

```python
m_car_df = pd.merge(brand_df,car_df,on='title_F',how='right')
```

```python
m_car_df.describe()
```

```python
(m_car_df.info())
```

```python
m_car_df.loc[:, 'Engine'] = m_car_df['Engine'].astype(str).str.replace('L', '', regex=False).astype(float)
```

```python
m_car_df[car_df['Engine'].isna()]
```

```python
m_car_df.loc[:, 'Engine'] = m_car_df.groupby('title')['Engine'].transform(lambda x : x.fillna(x.mode().iloc[0])if not x.mode().empty else x.fillna(car_df['Engine'].median()))
```

```python
m_car_df[m_car_df['Engine'].isna()]
```

```python
m_car_df.loc[:, 'Emission Class'] = m_car_df['Emission Class'].fillna(m_car_df['Emission Class'].mode()[0])
```

```python
m_car_df[m_car_df['Emission Class'].isna()]
```

```python
m_car_df.loc[:, 'Service history'] = m_car_df['Service history'].fillna('Unknown')
```

```python
m_car_df.info()
```

```python
car_df_dv = m_car_df.drop(['Previous Owners', 'Doors', 'Seats'], axis=1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(car_df_dv_num)
```

```python
import numpy as np
cumsum = np.cumsum(pca.explained_variance_ratio_)
```

```python
print(cumsum)
```

```python
n_components = np.argmax(cumsum >= 0.7) + 1
print(f" 70%      : {n_components}")
```

```python
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(car_df_dv_num)
```
